## Analyzing loss functions

### Mean Absolute Loss Function (L1 loss)

This loss function is used to minimize the absolute difference between predicted and actual values. This function is preferred when we want to be less strict towards outliers. On the downside, the gradient of the function is computationally expensive to compute, the derivative of the function at 0 is undefined, and the gradient is the same for each error.

### Quadratic Loss Function (L2 loss)

This loss function is used to minimize the squared difference between the predicted and the actual values. This is the most commonly used loss function in applications include linear regression. The gradient of the quadratic loss function is relatively inexpensive to calculate. Additionally, the quadratic loss function has a different gradient for each error which is helpful in optimization. The presence of outliers can have a large impact on models trained using this function.

### Huber Loss Function

The Huber loss function takes the form of the quadratic loss function at low residuals and the form of the absolute loss function at high residuals. It protects from high bias from outliers while also providing preferred characteristics of the quadratic function around the minimum. The concern in Huber loss is that the optimal choice of delta, or the point where the switch occurs, which is unknown.

## Code

```python
import numpy as np
import matplotlib.pyplot as plt

errors = np.linspace(-10, 10, 400)

absolute_loss = abs(errors)
quadratic_loss = 0.5*(errors)**2

def huber_loss_function (errors, delta) :
    cond = np.abs(errors) <= delta
    small_error_loss = 0.5*(errors)**2
    large_error_loss = delta * (np.abs(errors) - 0.5 * delta)
    return np.where(cond, small_error_loss, large_error_loss)

huber_loss = huber_loss_function(errors, 4)

plt.plot(errors, absolute_loss, label="Absolute loss")
plt.plot(errors, quadratic_loss, label="Quadratic loss")
plt.plot(errors, huber_loss, label="Huber loss")
plt.xlabel("Error")
plt.ylabel("Penalty")
plt.title("Loss functions")
plt.legend()
plt.show()
```
